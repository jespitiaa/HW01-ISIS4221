```python
relevance_query_1 = [0, 0, 0, 1]
#SOLUCION
def precision(relevance_query):
  sum=0
  for i in relevance_query:
    sum+=i
  return sum/len(relevance_query)

precision(relevance_query_1)
```

Precision at K (relevance is binary) LEIDY

```python
relevance_query_1 = [0, 0, 0, 1]
k = 1
#SOLUCION:
def precision_at_k(relevance_query, k1):
  sum=0
  for i in range(0,k1):
    sum+=relevance_query[i]
  return sum/k1

precision_at_k(relevance_query_1, k)
```

Recall at K (relevance is binary) ANAMARIA

```python
relevance_query_1 = [0, 0, 0, 1]
k = 1
number_relevant_docs = 4

def recall_at_k(relevance_query_1, number_relevant_docs, k):

  recall_k = sum(relevance_query_1[:k])/number_relevant_docs

  return recall_k  

recall_at_k(relevance_query_1, number_relevant_docs, k)
```

DCG at K (relevance is a natural number) MARIA

```python
relevance_query_3 = [4, 4, 3, 0, 0, 1, 3, 3, 3, 0]
k=6
#SOLUCION
def dcg_at_k(relevance_query, k):
  dcg=0
  for i in range(0, k):
    dcg+=relevance_query[i]/np.log2(np.max(np.array([i+1,2])))
  return dcg

dcg_at_k(relevance_query_3, k)
```
